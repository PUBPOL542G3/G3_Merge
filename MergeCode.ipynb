```python
#Import pandas library for data processing
import pandas as pd
```

```python
#Importing, reading, and displaying Mahnur's file from github
Mahnur_file_location = 'https://github.com/PUBPOL542G3/Mahnur-/raw/main/mahnurcleanedfile.xlsx'
DF_C02 = pd.read_excel(Mahnur_file_location)
display(DF_C02)
```

```python
#Importing, reading, and displaying Oliver's file from github
Oliver_file_location = 'https://github.com/PUBPOL542G3/Oliver/blob/main/OliverRevisedData.xlsx?raw=true'
DF_Employment = pd.read_excel(Oliver_file_location)
display(DF_Employment)
```

```python
#Importing, reading, and displaying Marina's file from github
Marina_file_location = 'https://github.com/PUBPOL542G3/Marina/blob/main/marina_cleaned.xlsx?raw=true'
DF_ForestArea = pd.read_excel(Marina_file_location)
display(DF_ForestArea)
```

```python
#Merging Oliver's cleaned employment data and Mahnur's cleaned CO2 data.  
Oliver_Nur_Merge = DF_Employment.merge(DF_C02,left_on='country',right_on='country',how='outer',indicator='check1')
display(Oliver_Nur_Merge)
```

```python
#Merging in Marina's cleaned data on forested area to Oliver and Mahnur's merged data.
Triple_Merge = Oliver_Nur_Merge.merge(DF_ForestArea,left_on='country',right_on='country',how='outer',indicator='check2')
```

```python
#displaying data to check that the merge worked.
display(Triple_Merge)
```

```python
#Importing, reading, and displaying data on country regions from github.
region_file_location = 'https://github.com/PUBPOL542G3/G3_Merge/blob/main/Regions%20(1).xlsx?raw=true'
DF_Regions = pd.read_excel(region_file_location,sheet_name='list-of-countries-etc')
display(DF_Regions)
```

```python
#refining data frame to only include the desired variables.
Refined_Col_Region = ['name', 'four_regions']
DF_RevRegions = DF_Regions[Refined_Col_Region]
display(DF_RevRegions)
```

```python
#Renaming the two variable columns in the refined data frame for regions so that we can merge the regions data with our data frame containing Co2, employment, and forested area data.
DF_RevRegions.rename(columns= {'name':'country'}, inplace= True)
DF_RevRegions.rename(columns= {'four_regions':'region'}, inplace= True)
display(DF_RevRegions)
```

```python
#renaming some of the country names in the regions data frame to match country names in our Triple_Merge data frame and facilitate merging the two data frames.
DF_FinRegions = DF_RevRegions.replace({
     'Congo, Dem. Rep.':'Democratic Republic of the Congo',
    'Congo, Rep.': 'Republic of the Congo',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Slovak Republic': 'Slovakia'
})
display(DF_FinRegions)
```

```python
#Merging redions data with Triple_Merge data.
DF_MergeRegion = DF_FinRegions.merge(Triple_Merge,left_on='country',right_on='country',how='left', indicator = 'True')
display(DF_MergeRegion)
```

```python
#Dropping the unwanted variables "check1", "check2", and "True", which were used to help confirm that our merged data frame contains null values. 
DF_MergeRegion.drop(['True', 'check1', 'check2'], 1, inplace = True)
display (DF_MergeRegion)
```

```python
#Here we are dropping any contries that have missing data (or an NaN value) for any of our variables. 
DF_MergeRegion.dropna(0,"any",inplace=True)

#We use the below code to display the full data frame. 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
display(DF_MergeRegion)
```
